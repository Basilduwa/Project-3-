```python
import os
import cv2
import numpy as np
from skimage.feature import hog

# Settings
image_size = (64, 64)
data_dir = './datasets/AugmentedFruitDataset/'

data, labels, class_names = [], [], []

# data -> your actual images in pixels
# label -> The labels of your images in number form
    # Apple -> 0
    # Banana -> 1
    # Orange -> 2
# class_names -> Your actual classnames
    # [Apple, Banana, Orange]

# loading images from our folders manually
for label, class_name in enumerate(sorted(os.listdir(data_dir))):
    class_path = os.path.join(data_dir, class_name)
    if os.path.isdir(class_path):
        class_names.append(class_name)
        for filename in os.listdir(class_path):
            if filename.lower().endswith(('.jpg', '.png', '.jpeg')):
                img_path = os.path.join(class_path, filename)
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                img = cv2.resize(img, image_size)

                features = hog(img, 
                orientations=9, 
                pixels_per_cell=(8, 8),
                cells_per_block=(2, 2), 
                block_norm='L2-Hys')
                
                data.append(features)
                labels.append(label)

# turning our arrays into numpy arrays for performance
data = np.array(data)
labels = np.array(labels)
```

```python
# 2.Picking your model type / architecture
from sklearn.model_selection import train_test_split

# Use data from the same distribution
# 95% Training / 5% Testing => 0.95Tra, 0.05Test ( a lot of data, a really large dataset )
# 70% Training / 30% Testing => 0.7Tra, 0.3Test ( medium dataset )
# 50% Training / 50% Testing => 0.5Tra, 0.5Test ( less data )

# 3.Split data
TrainData, TestData, TrainLabels, TestLabels = train_test_split(data, labels, test_size=0.3, random_state=42)
```

```python
# 4. Train Model
from sklearn import svm
model = svm.SVC(kernel='poly')
model.fit(TrainData, TrainLabels)
```

```python
# 5. Perform Metric Assessment

from sklearn.metrics import accuracy_score

# Test accuracy
TestPredictions = model.predict(TestData)

print("Accuracy:", accuracy_score(TestLabels, TestPredictions))
```

```python
# 6. Save Model

import joblib
# Save model and classes
joblib.dump(model, './models/svm/svm_model.joblib')
joblib.dump(class_names, './models/svm/svm_class_names.joblib')
```
